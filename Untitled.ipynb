```python
from typing import Union, TypeVar

_PANDAS_DF    = TypeVar('pandas.core.frame.DataFrame')
_FLOAT_OR_STR = Union[float, str]


def convert_alcohol_str(x: _FLOAT_OR_STR) -> float: 
    try:
        x = float(x)
    except ValueError:
        x = x[:2] + '.' + x[2]
        x = float(x)

    return x


wine_raw_df['alcohol'] = wine_raw_df['alcohol'].apply(convert_alcohol_str)
```

```python
wine_raw_df[wine_raw_df.alcohol > 20]
```

```python
def treat_alcohol_upper_outliers(x: float) -> float:
    if x > 70:
        x /= 10
        
    return x

wine_raw_df['alcohol'] = wine_raw_df['alcohol'].apply(treat_alcohol_upper_outliers)
```
